```python
def Model():
  models = []
  models.append(XGBClassifier(
      max_depth=5,objective='rank:pairwise',random_state=2,scale_pos_weight=0.25,n_estimators=100))
  models.append(LGBMClassifier(
      objective='binary',is_unbalance=True,metric='AUC',random_state=2,n_estimators=100
  ))
  return models
```

```python
def train(data,train_len):
    training = data[:train_len]
    testing = data[train_len:]

    train = training.drop(['Id','risk_flag'],axis=1)
    target = training['risk_flag']
    test = testing.drop(['Id','risk_flag'],axis=1)
    
    X_train,X_test,y_train,y_test = train_test_split(train,target,test_size=0.2,stratify=target,random_state=2)
    
    models = Model()
    for model in models:
      model.fit(X_train,y_train, eval_metric = 'auc', eval_set = [(X_test,y_test)])
      y_pred = model.predict(X_test)
      print(roc_auc_score(y_test,y_pred))
```

```python
def predictions(data,train_len,num):

  training = data[:train_len]
  testing = data[train_len:]

  train = training.drop(['Id','risk_flag'],axis=1)
  target = training['risk_flag']
  test_id = testing['Id']
  test = testing.drop(['Id','risk_flag'],axis=1)
  model = LGBMClassifier(
     objective='binary',is_unbalance=True,metric='AUC',random_state=2,n_estimators=300
  )
  model.fit(train,target)
  
  ans = model.predict(test)
  ans = ans.astype('int8')
  sub_df = pd.DataFrame({'id':test_id,'risk_flag':ans})
  sub_df.to_csv('sub{}.csv'.format(num),index=False)
```
